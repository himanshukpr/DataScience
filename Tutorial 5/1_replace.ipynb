```python
import pandas as pd
import numpy as np

df = pd.read_csv('data.csv')
new_df = df.replace({
    'Temperature (C)':'-999999',
    'Wind Speed (km/h)':'-999999',
    'Event':'No Event',
},np.nan)
new_df
```

```python
new_df = df.replace({
    '-999999': np.nan,
    'No Event': 'Sunny'
})
new_df
```

```python
new_df = df.replace({
    'Temperature (C)':'[A-Za-z]',
    'Wind Speed (km/h)':'[A-Za-z]'
},'',regex=True)
new_df
```
